```python
df= df.drop_duplicates()
```

```python
df.isnull().sum()
```

```python
# Fill with 'Unknown' for text columns
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
```

```python
# Handle missing date_added by filling with mode or dropna()
df['date_added'] = df['date_added'].fillna(df['date_added'].mode()[0])
```

```python
# Fill rating with mode
df['rating'] = df['rating'].fillna(df['rating'].mode()[0])
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df[['title', 'date_added']].head(10)
```

```python
print("Earliest Date:", df['date_added'].min())
print("Latest Date:", df['date_added'].max())
```

```python
# Optional: extract year and month for analysis
df['added_year'] = df['date_added'].dt.year
df['added_month'] = df['date_added'].dt.month
```

```python
print(df['type'].unique())
print(df['country'].unique())
```

```python
df['type'] = df['type'].str.lower().str.strip()
df['country'] = df['country'].str.lower().str.strip()
```

```python
print(df['type'].unique())
print(df['country'].unique())
```

```python
df['country'].apply(lambda x: x != x.strip()).sum()
```

```python
# Remove any entries that start with a comma
df['country'] = df['country'].str.lstrip(',').str.strip()

# Replace empty or malformed entries
df['country'] = df['country'].replace('', 'unknown')
```

```python
#save the cleaned dataset
df.to_csv("cleaned_netflix_data.csv", index=False)
```
